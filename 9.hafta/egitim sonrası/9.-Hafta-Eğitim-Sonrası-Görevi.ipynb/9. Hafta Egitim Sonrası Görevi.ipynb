# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durum = "Kapalı"
        self.tv_ses = 10
        self.kanal_listesi = ["TRT"]
        self.kanal = "TRT"

    def tv_ac(self):
        if self.tv_durum == "Açık":
            print("Televizyon zaten açık.")
        else:
            self.tv_durum = "Açık"
            print("Televizyon açılıyor...")
            time.sleep(1)
            print("Televizyon açık.")

    def tv_kapat(self):
        if self.tv_durum == "Kapalı":
            print("Televizyon zaten kapalı.")
        else:
            self.tv_durum = "Kapalı"
            print("Televizyon kapanıyor...")
            time.sleep(1)
            print("Televizyon kapalı.")

    def ses_ayarları(self):
        while True:
            işlem = input("Sesi azalt: '<'\nSesi artır: '>'\nÇıkış: 'q'\nSeçiminiz: ")
            if işlem == "<":
                if self.tv_ses > 0:
                    self.tv_ses -= 1
                    print(f"Ses: {self.tv_ses}")
                else:
                    print("Ses zaten en düşük seviyede.")
            elif işlem == ">":
                if self.tv_ses < 31:
                    self.tv_ses += 1
                    print(f"Ses: {self.tv_ses}")
                else:
                    print("Ses zaten en yüksek seviyede.")
            elif işlem == "q":
                break
            else:
                print("Geçersiz işlem.")

    def kanal_ekle(self):
        yeni_kanal = input("Eklemek istediğiniz kanalın adını girin: ")
        self.kanal_listesi.append(yeni_kanal)
        print(f"Kanal eklendi: {yeni_kanal}")

    def kanal_sayısı(self):
        print(f"Toplam kanal sayısı: {len(self.kanal_listesi)}")

    def rastgele_kanal(self):
        self.kanal = random.choice(self.kanal_listesi)
        print(f"Şu anki kanal: {self.kanal}")

    def tv_bilgileri(self):
        print("\nTelevizyon Bilgileri")
        print("--------------------")
        print(f"Durum: {self.tv_durum}")
        print(f"Ses: {self.tv_ses}")
        print(f"Kanal Listesi: {self.kanal_listesi}")
        print(f"Şu anki kanal: {self.kanal}")
        print("--------------------")

# Kullanıcı Etkileşimi
kumanda = Kumanda()

print("""
Kumanda Programına Hoşgeldiniz

İşlemler:
1 - Tv Aç
2 - Tv Kapat
3 - Ses Ayarları
4 - Kanal Ekle
5 - Kanal Sayısını Öğren
6 - Rastgele Kanala Geç
7 - Televizyon Bilgileri
Çıkmak için 'q' tuşuna basın.
""")

while True:
    işlem = input("Bir işlem seçiniz: ")

    if işlem == "q":
        print("Programdan çıkılıyor...")
        break
    elif işlem == "1":
        kumanda.tv_ac()
    elif işlem == "2":
        kumanda.tv_kapat()
    elif işlem == "3":
        kumanda.ses_ayarları()
    elif işlem == "4":
        kumanda.kanal_ekle()
    elif işlem == "5":
        kumanda.kanal_sayısı()
    elif işlem == "6":
        kumanda.rastgele_kanal()
    elif işlem == "7":
        kumanda.tv_bilgileri()
    else:
        print("Geçersiz işlem, tekrar deneyin.")
```
